```python
N=len(data['nodes'])
Nodes=data['nodes']
N
```

```python
L=len(data['links'])
Edges=[(data['links'][k]['source'], data['links'][k]['target']) for k in range(L)]

L
```

```python
g = nx.DiGraph()
g.add_nodes_from(data.keys())
g.add_edges_from(Edges)
```

```python
gg=g.to_undirected()
gg
#nx.Graph()
```

```python
gg.degree()
```

```python
for n, d in gg.degree():
    print('%s %d' % (n, d))
color = '#FF1493'
options = {
    'node_color': color ,
    'node_size': 5,
    'line_color': 'grey',
    'linewidths': 0,
    'width': 0.1,
}

nx.draw_random(gg, **options)
plt.show()
```
